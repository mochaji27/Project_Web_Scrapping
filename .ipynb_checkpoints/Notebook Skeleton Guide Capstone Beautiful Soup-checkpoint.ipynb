## Requesting the Data and Creating a BeautifulSoup

Let's begin with requesting the web from the site with `get` method.

```python
#Dari Halaman tersebut carilah judul , imdb rating , metascore, dan votes

import requests

url_get = requests.get('https://www.imdb.com/search/title/?release_date=2021-01-01,2021-12-31')
```

To visualize what exactly you get from the `request.get`, we can use .content so ee what we exactly get, in here i slice it so it won't make our screen full of the html we get from the page. You can delete the slicing if you want to see what we fully get.

```python
url_get.content[1:500]
```

As we can see we get a very unstructured and complex html, which actually contains the codes needed to show the webpages on your web browser. But we as human still confused what and where we can use that piece of code, so here where we use the beautifulsoup. Beautiful soup class will result a beautifulsoup object. Beautiful Soup transforms a complex HTML document into a complex tree of Python objects. 

Let's make Beautiful soup object and feel free to explore the object here.

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(url_get.content,"html.parser")
```

## Finding the right key to scrap the data & Extracting the right information

Find the key and put the key into the `.find()` Put all the exploring the right key at this cell. 
In this step, i using class 'lister-list' to get list of each box.

```python
soup.prettify()
```

```python
table = soup.find('div', attrs={'class':'lister-list'})
print(table.prettify()[1:500])
```

Finding row length.

```python
list = table.find_all('div', attrs={'class' : 'lister-item mode-advanced'})
row_length = len(list)
row_length
```

In this step, i has found class 'lister-item-header' to get movie name in tag 'a', class 'inline-block ratings-imdb-rating' to get movie rating and i using try with except for movie that not have rating score, class 'metascore mixed' to get meta score, and then class 'sort-num_votes-visible' to get votes bar


```python
temp = [] #initiating a tuple

for i in range(0, row_length):

    #scrapping process
    
    header = list[i].find('h3', attrs = {'class':'lister-item-header'})
    judul = header.find('a').text
    
    rating_bar = list[i].find(attrs = {'class':'inline-block ratings-imdb-rating'})
    #rating_bar.find('div')
    #rating_inline = rating_bar.find('div', attrs={'class':'inline-block ratings-imdb-rating'})

    #rating = rating_bar.find('strong')
    
    try:
        rating_list = list[i].find('div', attrs={'class':'inline-block ratings-imdb-rating'})
        rating = rating_list.find('strong').text.strip()
    except:
        rating = 0
    
    try:
        meta_score = list[i].find('span', {'class':'metascore mixed'}).text.strip()
    except:
        meta_score = 0
        
        
    votes_bar = list[i].find('p', {'class':'sort-num_votes-visible'})
    
    try:
        votes = votes_bar.find('span', attrs={'name':'nv'}).text.strip().replace(',', '.')
    except:
        votes = 0
        
        

    temp.append((judul, rating, meta_score, votes))
    
temp[1:5]
```

## Creating data frame & Data wrangling

Put the array into dataframe

```python
import pandas as pd

df = pd.DataFrame(temp, columns=('Tittle', 'Rating', 'MetaScore', 'Vote'))
df.head()
```

```python
df.dtypes
```

Change data type for some columns

```python
df['Rating'] = df['Rating'].astype('float64')
df['MetaScore'] = df['MetaScore'].astype('float64')
df['Vote'] = df['Vote'].astype('float64')
df = df.set_index('Tittle', drop=True)
df.head()
```

Sorting data by Vote column

```python
df = df.sort_values(by = 'Vote', ascending=False)
df.head(7)
```
